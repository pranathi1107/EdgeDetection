```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def perform_edge_detection(image_path):
    # Read the image
    original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply Gaussian blur to the image to reduce noise and help with edge detection
    blurred_image = cv2.GaussianBlur(original_image, (5, 5), 0)

    # Use the Canny edge detector
    edges = cv2.Canny(blurred_image, 50, 150)

    # Display the original and the edge-detected images side by side
    plt.subplot(121), plt.imshow(original_image, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(122), plt.imshow(edges, cmap='gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

    plt.show()

# Replace 'your_image_path.jpg' with the path to your image file
image_path = '/content/bird.jpg'
perform_edge_detection(image_path)
```
